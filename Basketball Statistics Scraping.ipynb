<h2 id="Section_1"> Define a Web Scraping Function  </h2>

In this section, I will define the function <code>get_basketball_stats</code>. The function will return a dictionary corresponding to the Regular season statistics of each basketball player scraped from m <a href="http://en.wikipedia.org">en.wikipedia.org</a>. 
The <a href="https://en.wikipedia.org/wiki/Parameter_(computer_programming)"> Formal Parameter </a> or input  <code> link</code> is going to be the URL of the Wikipedia page for each player. This function will return a  Python dictionary. The keys will be the column names, and the values will be a list that contains a different columns; the element of each list will be a separate row for that column. 

```python
import bs4
import requests
import pandas as pd
import numpy as np
# import boto3
```

```python
def get_basketball_stats(link='https://en.wikipedia.org/wiki/Michael_Jordan'):
    # read the webpage 
    response = requests.get(link)
    # create a BeautifulSoup object to parse the HTML  
    soup = bs4.BeautifulSoup(response.text, 'html.parser')
    # the player stats are defined  with the attribute CSS class set to 'wikitable sortable'; 
    #therefore I create a tag object "table"
    table=soup.find(class_='wikitable sortable')

    #the headers of the table are the first table row (tr) I create a tag object that has the first row  
    headers=table.tr
    #the table column names are displayed  as an abbreviation; therefore I find all the abbr tags and returns an Iterator
    titles=headers.find_all("abbr")

    #I create a dictionary  and pass the table headers as the keys 
    ordered_keys = []
    data = {}
    for title in titles:
        ordered_keys.append(title['title'])
        data[title['title']] = []

    #I will store each column as a list in a dictionary, the header of the column will be the dictionary key 

    #I iterate over each table row by fining each table tag tr and assign it to the objed
    for row in table.find_all('tr')[1:]:
        #I iterate over each cell in the table, as each cell corresponds to a different column I all obtain the correspondin key corresponding the columnn 

        #NB: Need to iterate through ordered keys
        for key,a in zip(ordered_keys,row.find_all("td")[2:]):
            # we append each elment and strip any extra HTML contnet 
            data[key].append(''.join(c for c in a.text if (c.isdigit() or c == ".")))

    # I remove extra rows by finding the smallest list     
    Min=min([len(x)  for x in data.values()])
    #I convert the elements in the key to floats 
    for key in data.keys():

        data[key]=list(map(lambda x: float(x), data[key][:Min]))

    return data
```

<h5>For each Player create a Python dictionary from the table <b>Regular season table </b>.</h5>
